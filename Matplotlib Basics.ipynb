# Stack Plots

```python
days = [1,2,3,4,5]

sleep = [6,7,8,8,7]
work = [8,8,8,8,10]
play = [3,3,3,2,5]
travel = [3,4,3,4,2]
other = [4,2,2,2,0]
```

```python
plt.stackplot(days,sleep,work,play,travel,other,colors=['c','m','g','b','k'])
plt.legend()
```

```python
plt.stackplot(days,sleep,work,play,travel,other,colors=['c','m','y','b','k'])
plt.plot([],color='c',label='sleep',linewidth=5)
plt.plot([],color='m',label='work',linewidth=5)
plt.plot([],color='y',label='play',linewidth=5)
plt.plot([],color='b',label='travel',linewidth=5)
plt.plot([],color='k',label='other',linewidth=5)
plt.legend()
```

# Subplots 

```python
#Figure, axes etc
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax1.plot(x,y)
ax1.set_title('First Plot')
ax2.plot(y,x)
ax2.set_title('Second Plot')

#Label Rotation
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(90)
    
#grid
ax2.grid(True,color='g',linestyle='--',linewidth=1)

#color labels
ax1.set_xlabel('My X label')
ax1.xaxis.label.set_color('green')

#ax3 = fig.add_subplot(2,1,2)
#ax3.plot(y,y)
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((6,2),(0,0),rowspan=2,colspan=2)
ax2 = plt.subplot2grid((6,2),(2,0),rowspan=2,colspan=1)
ax3 = plt.subplot2grid((6,2),(4,0),rowspan=2,colspan=2)
ax4 = plt.subplot2grid((6,2),(2,1),rowspan=2,colspan=1,sharey=ax2)
ax1.plot(x,y)
ax2.scatter(y,x)
ax3.plot(x,x)
ax4.scatter(y,y)
plt.setp(ax4.get_yticklabels(), visible=False)
```
